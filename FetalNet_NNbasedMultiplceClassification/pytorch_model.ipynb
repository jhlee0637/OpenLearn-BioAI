### Set Feature and Label Data

```python
# Set Feature data through dropping label column
X = df.drop('fetal_health', axis=1)

# Transform to array
X = X.values
X
```

* Feature data column: 21

```python
# Set label data
y = df['fetal_health']

# Transform
#     data type: float -> int
#     normal-suspect-pathological: 1-2-3 -> 0-1-2
y = df['fetal_health'].astype(int) - 1

# Transform to array
y = y.values
y
```

### Feature Engineering

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Divde dataset into training and test
from sklearn.model_selection import train_test_split

X_train, X_val, y_train, y_val = train_test_split(X_scaled, y, test_size=0.2)
```

```python
# check data type
print(len(X_train), X_train.dtype)
print(len(X_val), X_val.dtype)
print(len(y_train), y_train.dtype)
print(len(y_val), y_val.dtype)
```

* Data type will be transformed into float32 or int64
* Otherwise, errors would be raised while using pytorch

```python
# Transform data type
import torch
X_train_tensor = torch.tensor(X_train, dtype=torch.float32)
y_train_tensor = torch.tensor(y_train, dtype=torch.int64)

X_val_tensor = torch.tensor(X_val, dtype=torch.float32)
y_val_tensor = torch.tensor(y_val, dtype=torch.int64)
```

```python
# check data size
print(X_train_tensor.shape)
print(y_train_tensor.shape)
```

```python
# load data to TensorDataset
from torch.utils.data import TensorDataset
from torch.utils.data import DataLoader

# TensorDataset: indexed data
train_dataset = TensorDataset(X_train_tensor, y_train_tensor)
val_dataset = TensorDataset(X_val_tensor, y_val_tensor)
print(train_dataset[0]) #check

# Dataloader: non-indexed data batch
train_loader = DataLoader(train_dataset, batch_size=32)
val_loader = DataLoader(val_dataset, batch_size=32)
print(train_loader) #check

# optional: what if you REALLY want to see the DataLoader's data with indexing?
#it=iter(train_loader)
#print(len(next(it)[1]))
#next(it)[0]
```

## Neural Network Model for Multiclassification Problem

### Build Neural Network Model

```python
import torch
import torch.nn as nn

class FetalNet(nn.Module):
    def __init__(self):
        super().__init__()
        self.fc1 = nn.Linear(21, 64) #21 feature
        self.fc2 = nn.Linear(64,64)
        self.fc3 = nn.Linear(64,3) #3 classifications 
        self.relu = nn.ReLU()

    def forward(self, x): #forward propagation
        x = self.fc1(x)
        x = self.relu(x)
        x = self.fc2(x)
        x = self.relu(x)
        x = self.fc3(x)
        #skip softmax function: CrossEntropy in the next cell including it
        return x
```

```python
model = FetalNet()
model
```

### Train Neural Network Model

```python
from torch import Tensor
import torch.optim as optim

# Loss function
criterion = nn.CrossEntropyLoss() #softmax + NLLLoss

# Optimizer: Adam
optimizer = optim.Adam(model.parameters(), lr=1e-3) #learning rate: 0.001

num_epochs = 1000 #1 epoch = fully training one dataset
for epoch in range(num_epochs):
    model.train() #set training mode
    cost = 0 #set cost as zero for each epochs

    #training in batch size
    for xb, yb in train_loader: #data batch
        pred = model(xb) #forward propagation
        loss: Tensor = criterion(pred, yb) #loss value
        
        optimizer.zero_grad() #reset a gradient

        loss.backward() #back propagation
        optimizer.step() #update parameters

        cost += loss.item() 

    model.eval() #set evaluation mode

    correct = 0 
    sample_total = 0

    with torch.no_grad(): #prevent gradient calculation whild excuting the evaluation mode
        for xb, yb in val_loader: #test dataset
            pred = model(xb) #prediction
            _, predicted = torch.max(pred, 1) #choose most high possibility among 0-1-2 class
            correct += (predicted==yb).sum().item() #if correct, sum
            sample_total += yb.size(0)

    acc = correct / sample_total * 100

    if epoch%100==0:
        print(f"Epoch {epoch+1}/{num_epochs}, total_loss:{cost:.2}, Accuracy:{acc:.2}")

print(f"Epoch {epoch+1}/{num_epochs}, total_loss:{cost:.2}, Accuracy:{acc:.2}")
```
